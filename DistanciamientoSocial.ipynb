# S.I.R

una estrategia empleada para modelar enfermedades contagiosas, que pueden propagarse entre los miembros de una poblacin, es el modelo comportamental. Este tipo de modelo separa una poblacin en diferentes comportamientos. La idea es obtener informacion acerca de qu tan rpido sucede dicha propagacin, qu proporcin de la poblacin se infecta, qu proporcin morira, etc. Diferentes comportamientos se pueden indicar como (modelo bsico):

- **S**usceptible (an puede estar infectado, "sano")


- **I**nfectado


- **R**ecuperado (ha sido infectado y no puede volver a infectarse)

Para definir las variables involucradas en este simple modelo veamos el siguiente ejemplo.

## Ejemplo

Supngase que se tiene una poblacin de $1000$ habitantes ($N=1000$) y se sabe que $400$ de ellas se han infectado en un tiempo, $t$, de $7$ das despus del brote de la epidemia. Lo anterior se puede denotar como $S(7)=400$. En esta nueva enfermedad, la probabilidad de que una persona infectada infecte a una persona sana es del $20\%$, y supongamos que el nmero promedio de personas con las que una persona est en contacto por da es de $5$. Entonces, por da, un individuo infectado se encuentra con $5$ personas resultando infectadas el $20\%$ de ellas, es decir, la persona enferma infectar a $1$ persona por da. Se denota a esto con el parmetro $\beta$, que es la cantidad esperada de personas que una persona infectada infecta por da.

Sea $D$ la cantidad de das que una persona infectada tiene y puede transmitir la enfermedad, este dato es extremadamente importante. Si $D = 7$, una persona infectada camina durante siete das propagando la enfermedad e infecta a $1$ persona por da ($\beta = 1$). Entonces, esperamos que una persona infectada infecte a $7$ personas ms. Este es el nmero de reproduccin bsico, $R_0$, que representa el nmero total de personas que infecta una persona infectada, es decir, $R_0=\beta \times D$.

Si se piensa en $D$ como el nmero de das que una persona infectada tiene la enfermedad, entonces se puede pensar en otro parmetro $\gamma$, como la tasa de recuperacin, o la proporcin de infectados que se recuperan por da. Por ejemplo, si se tienen actualmente $30$ personas infectadas y $D=3$ (por lo que estn infectadas durante tres das), entonces, por da $1/3$ de ellas se recuperarn ($10$ personas), es decir, $\gamma=1/3$. Siendo $\gamma=1/D$, entonces $D=1/\gamma$ y $R_0=\beta \times D$, se deduce que $R_0=\beta / \gamma$.

Resumiendo, las variables involucradas en el ejemplo anterior son:

- $N$: Total de la poblacin


- $S(t)$: Nmero de personas susceptibles de ser infectadas en un da $t$


- $I(t)$: Nmero de personas infectadas en un da $t$


- $R(t)$: Nmero de personas recuperadas en un da $t$



- $\beta$: Cantidad esperada de personas que una persona infectada infecta por da


- $D$: Nmero de das que una persona infectada tiene y puede transmitir la enfermedad


- $\gamma$: Proporcin de infectados recuperados por da ($\gamma = 1/D$)



- $R_0$: Nmero total de personas infectadas por una persona infectada ($R_0=\beta / \gamma$)




# Derivacin de las ecuaciones

Deseamos determinar el nmero de *infectados*, *susceptibles* y *recuperados* para todos los das, dados nicamente los parmetros $\gamma$, $\beta$ y $N$. Es difcil obtener una frmula directa para $S(t)$, $I(t)$ y $R(t)$, sin embargo, es muy simple describir el cambio por da de $S$, $I$ y $R$, es decir, como cambia el nmero de *susceptibles* / *infectados* / *recuperados* dependiendo de los actuales valores.

Ahora estamos en el da $t$ despus del brote de la enfermedad. An as, la cantidad esperada de personas que una persona infectada infecta por da es $1$ ($\beta = 1$) y el nmero de das que una persona infectada tiene y puede transmitir la enfermedad es $7$ ($\gamma =1/7$ y $D=7$).

Digamos que en el da $t$, $60$ personas estn infectadas, $I (t) = 60$, la poblacin total es $100$, $N = 100$, y $30$ personas an son susceptibles, $S(t) = 30$ y $R( t) = 1006030 = 10$. Ahora, cmo cambian $S(t)$, $I(t)$ y $R(t)$ al da siguiente?

Se tienen $60$ personas infectadas. Cada uno de ellos infecta a $\beta = 1$ persona por da. Sin embargo, solo $30/100 = 30\%$ de las personas conocidas todava son susceptibles y pueden infectarse (eso es $S(t) / N$). Entonces, infectan a $60 \times 1 \times 30/100 = 18$ personas (de nuevo, piensen en ello hasta que tenga sentido: $60$ infectados que infectan en promedio $1$ persona por da, pero solo $30$ de $100$ personas an pueden estar infectadas, por lo que no infecte a $60 \times 1$ personas, sino que solo a $60 \times 1 \times 30 / 100 = 18$ personas). Entonces, $18$ personas de los susceptibles se infectan, por lo que $S(t)$ cambia en menos $18$. Al conectar las variables, acabamos de derivar la primera frmula:

$$\text{cambio de } S(t) \text{ al da siguiente } = -\beta \cdot I(t) \cdot S(t) / N$$

Lo anterior es la expresin de la deriva, $S'(t) = dS/dt$, y reescribiendo

$$S'(t) = dS(t)/dt = -\beta \cdot I(t) \cdot S(t) / N$$

Ahora, cmo cambia la cantidad de infectados? hay algunas personas nuevas infectadas. Exactamente la cantidad de personas que "*salen*" de $S(t)$ "*llegan*" a $I(t)$. Entonces, tenemos $18$ nuevos infectados y ya sabemos que la frmula ser similar a esta: $I'(t)= +  \times I(t) \times S(t)/N$ (se gana la cantidad exacta que $S(t)$ pierde, por lo que simplementese cambia el signo). Solo falta una cosa: algunas personas se recuperan. Recuerde, tenemos $\gamma$ para eso, es la proporcin de infectados que se recuperan por da, eso es justo lo que necesitamos!

Tenemos $60$ infectados y $\gamma = 1/3$, por lo que un tercio de los $60$ se recupera. Eso es $1/3 \times 60 = 20$. Finalmente, obtenemos la frmula:

$$I'(t) = dI(t)/dt = \beta \cdot I(t) \cdot S(t) / N - \gamma \cdot I(t)$$

El primer trmino del lado derecho de la ecuacin es el recin infectado de los susceptibles. La segunda parte son las recuperaciones.

Finalmente, llegamos a la ltima frmula, el cambio en las recuperaciones. Los recin recuperados son exactamente los $20$ que acabamos de calcular; no hay personas saliendo del compartimento "*recuperado*". Se asume que una vez recuperados, permanecen inmunes:

$$R'(t) = dR(t)/dt = \gamma \cdot I(t)$$

Ya contamos con las frmulas que se buscaban:

$$
\begin{align}
\frac{dS}{dt} & = -\beta \cdot I \cdot \frac{S}{N} \\
\frac{dI}{dt} & = \beta \cdot I \cdot \frac{S}{N} - \gamma \cdot I \\
\frac{dR}{dt} & = \gamma \cdot I
\end{align}
$$

Este es un conjunto de Ecuaciones Diferenciales Ordinarias (ODEs).

Ahora podemos describir el cambio en el nmero de personas *susceptibles*, *infectadas* y *recuperadas*. A partir de estas frmulas, afortunadamente, podemos calcular los nmeros que realmente nos interesan: $S(t)$, $I(t)$ y $R(t)$, el nmero de personas *susceptibles*, *infectadas* y *recuperadas* por cada da $t$.

# Codificando el modelo

Ahora vamos a codificar las ecuaciones. Se emplearn los cdigos implementados en [Infectious Disease Modelling: Understanding the models that are used to model Coronavirus](https://towardsdatascience.com/infectious-disease-modelling-part-i-understanding-sir-28d60e29fdfc): By Henri Froese, in Towardsdatascience.com, Apr 06, 2020. El cdigo completo lo pueden descargar de [su sitio en GitHub](https://github.com/hf2000510/infectious_disease_modelling).

Inicialmente se definiran los parmetros $N$, $\beta$, $D$, $\gamma$, y las condiciones iniciales, en $t=0$, para $S(t)$, $I(t)$ y  $R_0(t)$. 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
N = 1000
beta = 1.0  # infected person infects 1 other person per day
D = 4.0 # infections lasts four days
gamma = 1.0 / D

S0, I0, R0 = 999, 1, 0  # initial conditions: one infected, rest susceptible
```

Ahora se implementarn las frmulas determinadas

```python
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt
```

Ahora aqu es donde ocurre la magia: obtenemos nuestros valores $S(t)$, $I(t)$ y $R(t)$ de la funcin `odeint` de `Python` que toma las frmulas que definimos anteriormente, las condiciones iniciales y nuestras variables $N$, $\beta$ y $\gamma$, y calcula $S$, $I$ y $R$ para cada uno de los siguientes $50$ das.

```python
t = np.linspace(0, 50, 50) # Grid of time points (in days)
y0 = S0, I0, R0 # Initial conditions vector

# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T
```

```python
def plotsir(t, S, I, R):
    f, ax = plt.subplots(1,1,figsize=(10,4))
    ax.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
    ax.plot(t, I, 'y', alpha=0.7, linewidth=2, label='Infected')
    ax.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recovered')
    
    ax.set_xlabel('Time (days)')
    
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=True, which='major', c='w', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right', 'bottom', 'left'):
        ax.spines[spine].set_visible(False)
    plt.show();
    
```

```python
plotsir(t, S, I, R)
```

Como puede ver, solo toma alrededor de $30$ das para que casi una poblacin completa de $1000$ personas se infecte. Por supuesto, la enfermedad modelada aqu tiene un valor $R_0$ muy alto de $4.0$. Simplemente cambiando el nmero de personas que una persona infectada infecta por da $\beta$ a $0.5$ resulta en un escenario completamente diferente:

```python
N = 1000
beta = 0.50  # infected person infects 1 other person per day
D = 4.0 # infections lasts four days
gamma = 1.0 / D

S0, I0, R0 = 999, 1, 0  # initial conditions: one infected, rest susceptible
```

```python
t = np.linspace(0, 50, 50) # Grid of time points (in days)
y0 = S0, I0, R0 # Initial conditions vector

# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T
```

```python
plotsir(t, S, I, R)
```

Como puede ver, estos sistemas de EDO son extremadamente sensibles a los parmetros iniciales. Esa es tambin la razn por la cual es tan difcil modelar correctamente un brote emergente de una nueva enfermedad: simplemente no sabemos cules son los parmetros, e incluso los cambios leves producen resultados muy diferentes.

# *S.E.I.R*


En la literatura existen otros modelos ms completos, y complejos, como el SEIR, que es una extensin del S.I.R. y que considera, adems de los parmetros vistos, otros como:

entre muchos otros posibles parmetros adicionales, que podran complementar el modelo, pero lo haran mucho ms complejo.

Al derivar las ecuaciones, ya las consideramos intuitivamente como "*instrucciones*" que nos dicen lo que le sucede a la poblacin al da siguiente (por ejemplo, cuando $10$ personas estn infectadas y la recuperacin se produce a un ritmo de $\gamma = 1/5$, entonces el nmero de individuos recuperados al da siguiente debera aumentar en $1/5 \times 10 = 2$). Ahora solidificamos esta comprensin de las ecuaciones como "*direcciones*" o "*transiciones*" de un compartimento $S$, $I$ o $R$ a otro; esto simplificar enormemente las cosas cuando introduzcamos ms compartimentos ms adelante y las ecuaciones se vuelvan confusas.

# Introduciendo nuevos Compartimientos

## Obteniendo el compartimiento de *Expuestos*

Muchas enfermedades infecciosas tienen un perodo de incubacin antes de ser infecciosas durante el cual el husped an no puede transmitir la enfermedad. Llamaremos a tales individuos, y a todo el compartimento, *expuestos*.

Intuitivamente, tendremos transiciones de la forma $S \rightarrow E \rightarrow I \rightarrow R$: las personas susceptibles pueden contraer el virus y as quedar expuestos, luego infectados y luego recuperados. La nueva transicin $S \rightarrow E$ tendr la misma flecha que la transicin $S \rightarrow I$ actual, ya que la probabilidad es la misma (todos los *susceptibles* pueden estar *expuestos*), la tasa es la misma ("*exposicin*" ocurre inmediatamente) y la poblacin es igual (los individuos infecciosos pueden propagar la enfermedad y cada uno expone a $\beta$ nuevos individuos por da). Tampoco hay razn para que la transicin de $I$ a $R$ cambie. La nica nueva transicin es la de $E$ a $I$: la probabilidad es $1$ (todos los que estn expuestos se infectan), la poblacin es $E$ (todos los expuestos se infectarn) y la tasa obtiene una nueva variable, $\delta$(delta). Llegamos a estas transiciones:

<p float="left">
  <img src=https://miro.medium.com/max/1400/1*B-HZLpVWEgAZ3iloHBJjCA.png width="750" />
</p>



De estas transiciones, podemos derivar inmediatamente las nuevas ecuaciones:

$$
\begin{align}
\frac{dS}{dt} & = -\beta \cdot I \cdot \frac{S}{N} \\
\frac{dE}{dt} & = \beta \cdot I \cdot \frac{S}{N} - \delta \cdot E\\
\frac{dI}{dt} & = \delta \cdot E - \gamma \cdot I \\
\frac{dR}{dt} & = \gamma \cdot I
\end{align}
$$

## Programando el compartimiento de *Expuestos*

Retomando el cdigo realizado antes y cambiando algunas lneas para adicionarle los nuevos compartimientos. Como ejemplo se modelar una enfermedad altamente infecciosa, $R_0=5.0$ en una poblacin de $1$ milln, con un periodo de incubacin de $5$ das y una recuperacin de $7$ das.

```python
def plotsir(t, S, E, I, R):
    f, ax = plt.subplots(1,1,figsize=(10,4))
    ax.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
    ax.plot(t, E, 'y', alpha=0.7, linewidth=2, label='Expuesto')
    ax.plot(t, I, 'r', alpha=0.7, linewidth=2, label='Infectado')
    ax.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recuperado')
    
    ax.set_xlabel('Tiempo (dias)')
    ax.set_ylabel('Poblacin (habitantes)')
    
    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=True, which='major', c='w', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right', 'bottom', 'left'):
        ax.spines[spine].set_visible(False)
    plt.show();
```

Funcin para determinar las derivadas de las funciones

```python
def deriv(y, t, N, beta, gamma, delta):
    S, E, I, R = y
    dSdt = -beta * S * I / N
    dEdt = beta * S * I / N - delta * E
    dIdt = delta * E - gamma * I
    dRdt = gamma * I
    return dSdt, dEdt, dIdt, dRdt
```

Determinacin de los valores iniciales

```python
N = 1000000
D = 4.0 # infections lasts four days
gamma = 1.0 / D
delta = 1.0 / 5.0  # incubation period of three days
R_0 = 5.0
beta = R_0 * gamma # infected person infects 1 other person per day

S0, E0, I0, R0 = N-1, 1, 0, 0  # initial conditions: one exposed
```

Clculo de $S,E,I$ y $R$ para todo el tiempo de simulacin (100 das)

## Obteniendo el compartimiento de "*Decesos*"

Recordemos el modelo $S.E.I.R.$

Cundo puede morir una persona por la enfermedad? Slo mientras est infectado!, esto significa que hay qu agregar una transicin $I \rightarrow D$. 


Por supuesto, las personas no mueren de inmediato; Definimos una nueva variable $\rho$(rho) para la tasa a la que las personas mueren (por ejemplo, cuando tarda $6$ das en morir, $\rho=1/6$). No hay razn para que la tasa de recuperacin, $\gamma$, cambie. Entonces, nuestro nuevo modelo se ver:

<p float="left">
  <img src=https://miro.medium.com/max/1400/1*67WEZSCZ70oMs5TKKvvfJg.png width="500" />
</p>

Lo nico que falta son las probabilidades de pasar de infectado a recuperado y de infectado a muerto. Esa ser una variable ms (la ltima por ahora!), La tasa de mortalidad $\alpha$. Por ejemplo, si $\alpha= 5\%$, $\rho = 1$ y $\gamma=1$ (por lo que las personas mueren o se recuperan en $1$ da, eso es un ejemplo ms fcil) y $100$ personas estn infectadas, entonces $5\% \times 100 = 5$ personas morirn. Eso deja $95\% \times 100 = 95$ personas en recuperacin. En resumen, la probabilidad de $I \rightarrow D$ es $\alpha$ y, por lo tanto, la probabilidad de $I \rightarrow R$ es $1-\alpha$. Finalmente llegamos a este modelo:

<p float="left">
  <img src=https://miro.medium.com/max/1400/1*TIZaRpt70TR1RFtf2dmlew.png width="500" />
</p>

Adicionando esta nueva ecuacin al conjunto que tenamos de antes:

$$
\begin{align}
\frac{dS}{dt} & = -\beta \cdot I \cdot \frac{S}{N} \\
\frac{dE}{dt} & = \beta \cdot I \cdot \frac{S}{N} - \delta \cdot E\\
\frac{dI}{dt} & = \delta \cdot E - (1-\alpha) \cdot \gamma \cdot I - \alpha \cdot \rho \cdot I\\
\frac{dR}{dt} & = (1-\alpha) \cdot \gamma \cdot I \\
\frac{dD}{dt} & = \alpha \cdot \rho \cdot I
\end{align}
$$

## Programando el compartimiento de *Decesos*

Modificando el cdigo que venimos trayendo para incluir los decesos. Solo es necesario realizar algunos cambios menores, como es incluir $\alpha=20\%$ y $\rho=1/9$:

```python
def deriv(y, t, N, beta, gamma, delta, alpha, rho):
    S, E, I, R, D = y
    dSdt = -beta * S * I / N
    dEdt = beta * S * I / N - delta * E
    dIdt = delta * E - (1 - alpha) * gamma * I - alpha * rho * I
    dRdt = (1 - alpha) * gamma * I
    dDdt = alpha * rho * I
    return dSdt, dEdt, dIdt, dRdt, dDdt
```

```python
D = 4.0 # infections lasts four days
gamma = 1.0 / D
delta = 1.0 / 5.0  # incubation period of five days
R_0 = 5.0
beta = R_0 * gamma  # R_0 = beta / gamma, so beta = R_0 * gamma
alpha = 0.2  # 10% death rate
rho = 1/9  # 9 days from infection until death
S0, E0, I0, R0, D0 = N-1, 1, 0, 0, 0  # initial conditions: one exposed
```

```python
t = np.linspace(0, 100, 100) # Grid of time points (in days)
y0 = S0, E0, I0, R0, D0 # Initial conditions vector

# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma, delta, alpha, rho))
S, E, I, R, D = ret.T
```

```python
def plotseird(t, S, E, I, R, D):
    f, ax = plt.subplots(1,1,figsize=(10,4))
    ax.plot(t, S, 'b', alpha=0.7, linewidth=2, label='Susceptible')
    ax.plot(t, E, 'y', alpha=0.7, linewidth=2, label='Exposed')
    ax.plot(t, I, 'r', alpha=0.7, linewidth=2, label='Infected')
    ax.plot(t, R, 'g', alpha=0.7, linewidth=2, label='Recovered')
    ax.plot(t, D, 'k', alpha=0.7, linewidth=2, label='Dead')
    ax.plot(t, S+E+I+R+D, 'c--', alpha=0.7, linewidth=2, label='Total')

    ax.set_xlabel('Time (days)')

    ax.yaxis.set_tick_params(length=0)
    ax.xaxis.set_tick_params(length=0)
    ax.grid(b=True, which='major', c='w', lw=2, ls='-')
    legend = ax.legend()
    legend.get_frame().set_alpha(0.5)
    for spine in ('top', 'right', 'bottom', 'left'):
        ax.spines[spine].set_visible(False)
    plt.show();
```

```python
plotseird(t, S, E, I, R, D)
```

Se adicion una variable "*total*" que suma $S, E, I, R$ y $D$ para cada paso de tiempo como un "*control de sanidad*": los compartimentos siempre tienen que sumar $N$; Esto puede dar una pista sobre si las ecuaciones son correctas.

Ya sabemos cmo ir agregando nuevos compartimentos al modelo: piense qu transiciones deben agregarse y cambiarse; pensar en las probabilidades, poblaciones y tasas de estas nuevas transiciones; dibuja el diagrama; y finalmente escribe las ecuaciones. # Variables dependientes del tiempo

Resumamos el listado de las variables empleadas hasta ahora:

- $N$: poblacin total


- $S(t)$: nmero de personas susceptibles en el da $t$


- $E(t)$: nmero de personas expuestas el da $t$


- $I(t)$: nmero de personas infectadas el da $t$


- $R(t)$: nmero de personas recuperadas el da $t$


- $D(t)$: nmero de personas fallecidas el da $t$


- $\beta$: cantidad esperada de personas que una persona infectada infecta por da


- $D$: nmero de das que una persona infectada tiene y puede transmitir la enfermedad


- $\gamma$: la proporcin de infectados que se recuperan por da ($\gamma = 1/D$)


- $R_0$: el nmero total de personas que infecta una persona infectada ($R_0= \beta / \gamma$)


- $\delta$: duracin del perodo de incubacin


- $\alpha$: tasa de mortalidad


- $\rho$: tasa a la que muere la gente (= 1 / das desde la infeccin hasta la muerte)

Como puede ver, solo los compartimentos cambian con el tiempo (no son constantes). Por supuesto, esto es muy poco realista! Como ejemplo, por qu el valor $R_0$ debera ser constante? Seguramente, los bloqueos a nivel nacional reducen la cantidad de personas que infecta una persona infectada, de eso se trata! Naturalmente, para acercarnos a modelar desarrollos del mundo real, tenemos que hacer que nuestras variables cambien con el tiempo.
